# YOLO Part D: Oriented Bounding Box (OBB) Training

## Dataset: Aerial Objects OBB
- **Model:** YOLOv8 Nano OBB
- **Total Images:** 5
- **Training:** 4 images
- **Validation:** 1 image
- **Classes:** 5 (Model 1-5)
- **Type:** Rotated/Oriented Bounding Boxes

This notebook trains a YOLO OBB model to detect rotated objects with oriented bounding boxes.

⚠️ **Note:** This is a small demonstration dataset. Production systems typically use 100+ images.

## Step 1: Import Libraries and Setup

```python
from ultralytics import YOLO
import os
import sys
from pathlib import Path

# Create output directory
os.makedirs('models', exist_ok=True)

print("=" * 60)
print("PART D: ORIENTED BOUNDING BOX (OBB) MODEL TRAINING")
print("=" * 60)
print("\n📊 Dataset: Aerial Objects OBB (Rotated Detection)")
print("📦 Model: YOLOv8 Nano OBB")
print("💾 Output: models/obb_model.pt")
print("🔄 Detection Type: Rotated/Oriented Bounding Boxes")
print("\n⚠️  NOTE: Small dataset (5 images total)")
print("   Training will be quick but watch for overfitting")
print("\n" + "=" * 60 + "\n")
```

## Step 2: Load Pretrained OBB Model

```python
print("[1/4] Loading YOLOv8 Nano OBB pretrained model...")
model = YOLO('yolov8n-obb.pt')  # nano OBB model
print("✅ Model loaded successfully!")
print(f"Model size: {model.model}")
```

## Step 3: Train the OBB Model

```python
print("\n[2/4] Starting training...")
print("⏱️  This may take 2-5 minutes (small dataset)")
print("-" * 60)

results = model.train(
    data='../Part_D_OBB/data.yaml',        # Path to OBB dataset yaml
    epochs=20,                              # Number of training epochs
    imgsz=640,                              # Image size
    batch=2,                                # Small batch for small dataset
    patience=5,                             # Early stopping patience
    device=0,                               # GPU device (0=first GPU, or 'cpu')
    save=True,                              # Save model after training
    verbose=True,                           # Verbose output
    workers=2,                              # Fewer workers for small dataset
    project='runs/obb',                     # Project folder
    name='obb_model_detection',             # Run name
    exist_ok=True                           # Overwrite existing
)

print("-" * 60)
print("✅ Training completed!")
```

## Step 4: Save the Model

```python
print("\n[3/4] Saving trained model...")
model.save('models/obb_model.pt')
print("✅ Model saved to: models/obb_model.pt")
print(f"File size: {os.path.getsize('models/obb_model.pt') / 1024 / 1024:.2f} MB")
```

## Step 5: Test OBB Inference with Rotated Boxes

```python
print("\n[4/4] Testing inference on sample image...")
test_images_path = '../Part_D_OBB/test/images'

if os.path.exists(test_images_path):
    test_images = [f for f in os.listdir(test_images_path) if f.endswith(('.jpg', '.png'))]
    if test_images:
        sample_image = os.path.join(test_images_path, test_images[0])
        print(f"Testing with: {test_images[0]}")
        
        results = model.predict(
            source=sample_image,
            save=True,
            conf=0.5,
            project='runs/obb',
            name='test_inference',
            exist_ok=True
        )
        print("✅ Inference completed! Results saved to runs/obb/test_inference/")
        print("   Visualization shows rotated bounding boxes")
    else:
        print("⚠️  No test images found")
else:
    print("⚠️  Test images folder not found, trying validation images...")
    val_path = '../Part_D_OBB/valid/images'
    if os.path.exists(val_path):
        val_images = [f for f in os.listdir(val_path) if f.endswith(('.jpg', '.png'))]
        if val_images:
            sample_image = os.path.join(val_path, val_images[0])
            print(f"Testing with validation image: {val_images[0]}")
            
            results = model.predict(
                source=sample_image,
                save=True,
                conf=0.5,
                project='runs/obb',
                name='test_inference',
                exist_ok=True
            )
            print("✅ Inference completed!")
```

## Step 6: OBB Classes Information

```python
print("\n🔄 OBB Classes (5):")
classes = ['Model 1', 'Model 2', 'Model 3', 'Model 4', 'Model 5']
for i, cls in enumerate(classes, 1):
    print(f"      {i}. {cls}")

print("\n📌 About OBB (Oriented Bounding Boxes):")
print("   - Detects objects with rotation angle")
print("   - Not axis-aligned like regular detection")
print("   - Common in aerial/satellite imagery")
print("   - Useful for rotated objects, ships, aircraft, etc.")
```
