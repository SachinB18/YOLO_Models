# YOLO Part A: Object Detection Training

## Dataset: Pothole Detection
- **Model:** YOLOv8 Nano
- **Total Images:** 2,742
- **Training:** 2,532 images
- **Validation:** 143 images
- **Test:** 67 images
- **Classes:** 1 (Pothole)

This notebook trains a YOLO model to detect potholes in road images.

## Step 1: Import Libraries and Setup

```python
from ultralytics import YOLO
import os
import sys
from pathlib import Path

# Create output directory
os.makedirs('models', exist_ok=True)

print("=" * 60)
print("PART A: OBJECT DETECTION MODEL TRAINING")
print("=" * 60)
print("\n📊 Dataset: Pothole Detection (via Roboflow)")
print("📦 Model: YOLOv8 Nano")
print("💾 Output: models/detection_model.pt")
print("\n" + "=" * 60 + "\n")
```

## Step 2: Load Pretrained Model

```python
print("[1/4] Loading YOLOv8 Nano pretrained model...")
model = YOLO('yolov8n.pt')  # nano model
print("✅ Model loaded successfully!")
print(f"Model size: {model.model}")
```

## Step 3: Train the Model

```python
print("\n[2/4] Starting training on GPU...")
print("🚀 Using NVIDIA GeForce RTX 4060 for fast training")
print("-" * 60)

results = model.train(
    data='data.yaml',                       # Path to dataset yaml
    epochs=10,                              # Number of training epochs
    imgsz=640,                              # Image size
    batch=16,                               # Batch size (GPU optimized)
    patience=5,                             # Early stopping patience
    device=0,                               # GPU device (0 = first GPU)
    save=True,                              # Save model after training
    verbose=True,                           # Verbose output
    workers=8,                              # Number of workers (GPU optimized)
    project='runs/detect',                  # Project folder
    name='pothole_detection',               # Run name
    exist_ok=True                           # Overwrite existing
)

print("-" * 60)
print("✅ Training completed!")
```

## Step 4: Save the Model

```python
print("\n[3/4] Saving trained model...")
model.save('models/detection_model.pt')
print("✅ Model saved to: models/detection_model.pt")
print(f"File size: {os.path.getsize('models/detection_model.pt') / 1024 / 1024:.2f} MB")
```

## Step 5: Test Inference on Sample Image

```python
print("\n[4/4] Testing inference on sample image...")
test_images_path = '../Part_A_Detection/test/images'

if os.path.exists(test_images_path):
    test_images = [f for f in os.listdir(test_images_path) if f.endswith(('.jpg', '.png'))]
    if test_images:
        sample_image = os.path.join(test_images_path, test_images[0])
        print(f"Testing with: {test_images[0]}")
        
        results = model.predict(
            source=sample_image,
            save=True,
            conf=0.5,
            project='runs/detect',
            name='test_inference',
            exist_ok=True
        )
        print("✅ Inference completed! Results saved to runs/detect/test_inference/")
    else:
        print("⚠️  No test images found")
else:
    print("⚠️  Test images folder not found")
```

## Step 6: Summary

```python
print("\n" + "=" * 60)
print("✅ DETECTION MODEL TRAINING COMPLETED")
print("=" * 60)
print("\n📁 Output Files:")
print(f"   - Model: models/detection_model.pt")
print(f"   - Runs: runs/detect/pothole_detection/")
print(f"\n🎯 Metrics saved in: runs/detect/pothole_detection/results.csv")
print("\n" + "=" * 60 + "\n")
```
